```python
from qiskit.quantum_info import Statevector
import numpy as np
from numpy import pi, arccos, arctan2, sqrt

# Crear circuito y estado
circ = qiskit.QuantumCircuit(1,1)
circ.ry(pi/4, 0)
estado = Statevector.from_instruction(circ)

# Obtener amplitudes
alpha, beta = estado.data

# Coordenadas cartesianas de Bloch
x = 2 * (alpha.conjugate() * beta).real
y = 2 * (alpha.conjugate() * beta).imag
z = abs(alpha)**2 - abs(beta)**2

# Coordenadas esféricas
r = sqrt(x**2 + y**2 + z**2)
theta = arccos(z / r)
phi = arctan2(y, x)

print(f"Coordenadas esféricas:")
print(f"  r     = {r:.4f}")
print(f"  θ (theta) = {theta:.4f} rad ({np.degrees(theta):.2f}°)")
print(f"  φ (phi)   = {phi:.4f} rad ({np.degrees(phi):.2f}°)")
```

```python


import qiskit
from qiskit import QuantumCircuit
from qiskit.visualization import plot_histogram
from qiskit.visualization import plot_bloch_multivector
from qiskit.visualization import plot_bloch_vector
import pylatexenc
from qiskit_aer import AerSimulator
from qiskit.quantum_info import Statevector


import matplotlib.pyplot as plt
from numpy import pi
```

```python
# Creamos un circuito de 4 qubits y 4 bits clasicos
qc = QuantumCircuit(4, 4)

# 1) Hadamard en q0
qc.h(0)

# 2) NOT (X) en q1 y q2
qc.x(1)
qc.x(2)

# 3) CNOT sobre q2, con control q0
qc.cx(0, 2)

# 4) Toffoli (CCX) sobre q3, con controles q0,q1
qc.ccx(0, 1, 3)

# 5) Fredkin (CSWAP) entre q2 y q3, con control q0
qc.cswap(0, 2, 3)

# 6) Medida de todos los qubits
qc.measure([0, 1, 2, 3], [0, 1, 2, 3])

# Mostrar el circuito
display(qc.draw('mpl',style='textbook'))

import qiskit.providers.basic_provider

provider = qiskit.providers.basic_provider.BasicProvider()
backend = provider.get_backend()
qct2 = qiskit.transpile(qc, backend)
job = backend.run(qct2, shots=10000)
result = job.result()
counts = result.get_counts()
print(counts)
histograma = qiskit.visualization.plot_histogram(counts)
histograma
```

Código del algoritmo de Deutsch

```python
import qiskit
import pylatexenc
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt
```

```python
# Implementación de U_fc1 como oráculo de la función constante f(0)=f(1)=0
circ1 = qiskit.QuantumCircuit(2)
gatefc1 = circ1.to_gate(label='U_fc1')

# Implementación de U_fc2 como oráculo de la función constante f(0)=f(1)=1
circ2 = qiskit.QuantumCircuit(2)
circ2.x(1)  # y -> y ⊕ 1
gatefc2 = circ2.to_gate(label='U_fc2')

# Implementación de U_fb1 como oráculo de la función balanceada f(0)=0, f(1)=1
circ3 = qiskit.QuantumCircuit(2)
circ3.cx(0, 1) # y -> y ⊕ x
gatefb1 = circ3.to_gate(label='U_fb1')

# Implementación de U_fb2 como oráculo de la función balanceada f(0)=1, f(1)=0
circ4 = qiskit.QuantumCircuit(2)
circ4.x(1)       # y -> y ⊕ 1
circ4.cx(0, 1)   # y -> (y ⊕ 1) ⊕ x
gatefb2 = circ4.to_gate(label='U_fb2')

# Algoritmo de Deutsch
def deutsch_algorithm(gate):
  djf = qiskit.QuantumCircuit(2,1)
  djf.x(1)
  djf.barrier()
  djf.h(0)
  djf.h(1)

  djf.append(gate, [0,1], [])
  djf.h(0)

  djf.barrier()
  djf.measure(0, 0)
  display(djf.draw('mpl',style='textbook'))


  simulator = AerSimulator()
  qct2 = qiskit.transpile(djf, simulator)
  job = simulator.run(qct2, shots=1024)
  result = job.result()
  counts = result.get_counts()
  print(counts)
  display(qiskit.visualization.plot_histogram(counts))
```

```python
deutsch_algorithm(gatefc1)
```

```python
deutsch_algorithm(gatefc2)
```

```python
deutsch_algorithm(gatefb1)
```

```python
deutsch_algorithm(gatefb2)
```
